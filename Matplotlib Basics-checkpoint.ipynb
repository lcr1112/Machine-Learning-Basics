```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
import numpy as np
```

```python
x=np.arange(3,15,2)
```

```python
y=np.arange(2,20,3)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('2-D Plot')
```

```python
a=x*x*x
```

```python
a
```

```python
plt.plot(x,a,'r*',linestyle='--')
```

```python
b=pow(x,5)
```

```python
b
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'g*','r')
plt.subplot(2,2,2)
plt.plot(x,a,'r--','r')
plt.subplot(2,2,3)
plt.plot(x,b,'b+','b')
plt.subplot(2,2,4)
plt.plot(a,b,'y--','y')
plt.tight_layout()
```

```python
np.pi
```

```python
p=np.arange(0,6*np.pi,0.1)
```

```python
q=np.sin(p)
s=np.cos(p)
```

```python
plt.subplot(2,1,1)
plt.plot(p,q,'r--')
plt.title('Sine')
plt.subplot(2,1,2)
plt.plot(p,s,'g--')
plt.title('Cosine')
plt.tight_layout()
```

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True)

plt.axis('equal')
plt.show()
```
